```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from tqdm import tqdm
```

```python
df_train=pd.read_csv("train.csv",parse_dates=['date'], index_col='date')
df_test=pd.read_csv("test.csv",parse_dates=['date'], index_col='date')
df_train['month']=df_train.index.month
df_train['day']=df_train.index.day
df_test['month']=df_test.index.month
df_test['day']=df_test.index.day

df_test['otv']=np.zeros(len(df_test))
```

```python
df_test.head(5)
```

```python
for i in tqdm(range(len(df_test))):
    tek_prod_id=df_test.iloc[i,:]['product_id']
    tek_store_id=df_test.iloc[i,:]['warehouse_id']
    tek_day=df_test.iloc[i,:]['day']
    
    tek_df=df_train[df_train['product_id']==tek_prod_id]
    tek_df=tek_df[tek_df['warehouse_id']==tek_store_id]
    
    tek_df1=tek_df[tek_df['day']==tek_day]
    if(len(tek_df1)>0):
        df_test.iloc[i,5]=np.round(tek_df1['quantity'].mean())
    else:
        tek_df2=df_train[df_train['day'].isin([tek_day-1,tek_day+1]) ]
        if (len(tek_df2)>0):
            df_test.iloc[i,5]=np.round(tek_df2['quantity'].mean())
        else:
            df_test.iloc[i,5]=0.0
```

```python
df_test.head(5)
```

```python
new_df=df_test.copy()
new_df['quantity']=df_test['otv'].astype(int)
new_df.drop(columns=['product_id','warehouse_id','month','day','otv'],inplace=True)
new_df.to_csv("y_predTest.csv",index=False)
```
